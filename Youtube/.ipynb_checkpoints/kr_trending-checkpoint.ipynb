```python
import pandas as pd
```

```python
kr_youtube_trending_data = pd.read_csv('KR_youtube_trending_data.csv/KR_youtube_trending_data.csv')
```

```python
import json
# Load the category ID JSON file
with open("KR_category_id.json", 'r') as f:
    category_data = json.load(f)

# Create a dictionary to map category IDs to category titles
category_mapping = {item['id']: item['snippet']['title'] for item in category_data['items']}

# Map the category IDs to category titles in the dataframe
kr_youtube_trending_data['category'] = kr_youtube_trending_data['categoryId'].astype(str).map(category_mapping)

# Check the updated dataframe
print(kr_youtube_trending_data[['categoryId', 'category']].head())
```

```python
kr_youtube_trending_data.head()
```

```python
# Convert relevant columns to appropriate data types
kr_youtube_trending_data['publishedAt'] = pd.to_datetime(kr_youtube_trending_data['publishedAt'])
kr_youtube_trending_data['trending_date'] = pd.to_datetime(kr_youtube_trending_data['trending_date'])

# Select relevant columns for correlation analysis
kr_youtube_trending_data_cleaned = kr_youtube_trending_data[['view_count', 'likes', 'dislikes', 'comment_count']]

# Calculate correlation matrix
correlation_matrix = kr_youtube_trending_data_cleaned.corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, annot_kws={"size": 12})
plt.title('Correlation Matrix of KR YouTube Trending Data')
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```

```python
# Plot the number of trending videos per category over time
plt.figure(figsize=(15, 10))
category_over_time = kr_youtube_trending_data.groupby(['year_month', 'category']).size().unstack().fillna(0)

# Create a separate line plot for each category
for category in category_over_time.columns:
    plt.plot(category_over_time.index.to_timestamp(), category_over_time[category], label=category)

plt.title('Number of Trending Videos per Category Over Time')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Number of Videos')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python

# Calculate the correlation between category and view_count
correlation = kr_youtube_trending_data[['categoryId', 'view_count']].corr()

# Display the correlation matrix
print(correlation)

# Visualize the correlation
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Between Category and Number of Views')
plt.show()
```

```python

# Select relevant columns for correlation analysis
columns_of_interest = ['categoryId', 'view_count', 'likes', 'comment_count']
kr_youtube_trending_data_subset = kr_youtube_trending_data[columns_of_interest]

# Convert categoryId to numeric if it's not already
kr_youtube_trending_data_subset['categoryId'] = pd.to_numeric(kr_youtube_trending_data_subset['categoryId'], errors='coerce')

# Calculate the correlation matrix
correlation_matrix = kr_youtube_trending_data_subset.corr()

# Print the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix of categoryId, view_count, likes, and comment_count')
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```
