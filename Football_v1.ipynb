```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('final_data.csv')
```

```python
data = df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['team_encoded'] = label_encoder.fit_transform(data['team'])
```

```python
from scipy.stats import f_oneway

def perform_anova(data,column,target = 'current_value'):
    df_anova = data[[data[column].name,target]]
    grouped_data = [group[target] for _, group in df_anova.groupby(data[column].name)]
    return f_oneway(*grouped_data)

f_statistic, p_value = perform_anova(data,'team_encoded')
print(f"Anova Result: F statistic = {f_statistic}, P value = {p_value}")
```
