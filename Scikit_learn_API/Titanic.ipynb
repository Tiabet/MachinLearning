```python
from sklearn.preprocessing import OneHotEncoder

def encode_feature(df):
    features = ['Cabin','Sex','Embarked']
    for feature in features :
        oh_encoder = OneHotEncoder()
        df_feature = np.array(df[feature]).reshape(-1,1)
        oh_encoder = oh_encoder.fit(df_feature)
        df[feature] = oh_encoder.transform(df_feature).toarray()
        
    return df
```

```python
oh_encoder = OneHotEncoder()
df_feature = np.array(df['Sex']).reshape(-1,1)
oh_encoder = oh_encoder.fit(df_feature)
df_sex = oh_encoder.transform(df_feature).toarray()
```

```python
df[['Sex0','Sex1']] = pd.DataFrame(df_sex)
```
